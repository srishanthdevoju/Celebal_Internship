# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
import logging

logging.basicConfig(level=logging.INFO)
````

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
import re

def agent(query):
    logging.info(f"Received Query: {query}")
    query_lower = query.lower()

    # 🧠 Level-Up: Trigger Word Lists
    math_triggers = ["calculate", "math", "add", "multiply", "solve"]
    keyword_triggers = ["keywords", "extract", "summarize"]

    try:
        # Route 1: Math queries
        if any(trigger in query_lower for trigger in math_triggers):
            logging.info("Routing to Calculator")

            # 🧹 Level-Up: Regex to strip everything except numbers and math operators
            expression = re.sub(r'[^0-9+\-*/().]', '', query_lower)

            result = calculator(expression)

            if result == "Error in calculation":
                return {
                    "type": "error",
                    "result": "Invalid mathematical expression."
                }
            return {
                "type": "calculation",
                "result": result
            }

        # Route 2: Keyword queries
        elif any(trigger in query_lower for trigger in keyword_triggers):
            logging.info("Routing to Keyword Tool")

            # 🧹 Clean up the text by removing the trigger words
            text = query_lower
            for trigger in keyword_triggers:
                text = text.replace(trigger, "")
            # Remove common connecting words
            text = text.replace("from", "").replace("the", "").strip()

            result = extract_keywords(text)
            return {
                "type": "keywords",
                "result": result
            }

        # Route 3: General queries
        else:
            logging.info("Routing to General Response")
            return {
                "type": "general",
                "result": f"I received your general query: {query}"
            }

    except Exception as e:
        return {
            "type": "error",
            "result": str(e)
        }
````

  "result": ...

````python
import random
from datetime import datetime

# 🛠️ TOOL 3: Coin Flipper
def flip_coin() -> str:
    """Simulates a coin flip."""
    try:
        return random.choice(["Heads", "Tails"])
    except Exception:
        return "Error flipping coin"
````

````python
# 🛠️ TOOL 4: Time Teller
def get_current_time() -> str:
    """Returns the current date and time."""
    try:
        return datetime.now().strftime("%I:%M %p on %B %d, %Y")
    except Exception:
        return "Error getting time"
````

````python
import re

def agent(query):
    logging.info(f"Received Query: {query}")
    query_lower = query.lower()

    # 🧠 Level-Up: Expanded Trigger Word Lists
    math_triggers = ["calculate", "math", "add", "multiply", "solve"]
    keyword_triggers = ["keywords", "extract", "summarize"]
    coin_triggers = ["flip", "coin", "toss"]
    time_triggers = ["time", "date", "clock"]

    try:
        # Route 1: Math queries
        if any(trigger in query_lower for trigger in math_triggers):
            logging.info("Routing to Calculator")
            expression = re.sub(r'[^0-9+\-*/().]', '', query_lower)
            result = calculator(expression)

            if result == "Error in calculation":
                return {"type": "error", "result": "Invalid mathematical expression."}
            return {"type": "calculation", "result": result}

        # Route 2: Keyword queries
        elif any(trigger in query_lower for trigger in keyword_triggers):
            logging.info("Routing to Keyword Tool")
            text = query_lower
            for trigger in keyword_triggers:
                text = text.replace(trigger, "")
            text = text.replace("from", "").replace("the", "").strip()

            result = extract_keywords(text)
            return {"type": "keywords", "result": result}

        # Route 3: Coin Flip queries (NEW)
        elif any(trigger in query_lower for trigger in coin_triggers):
            logging.info("Routing to Coin Flipper")
            result = flip_coin()
            return {"type": "coin_flip", "result": result}

        # Route 4: Time queries (NEW)
        elif any(trigger in query_lower for trigger in time_triggers):
            logging.info("Routing to Time Teller")
            result = get_current_time()
            return {"type": "time", "result": result}

        # Route 5: General queries (Fallback)
        else:
            logging.info("Routing to General Response")
            return {"type": "general", "result": f"I received your general query: {query}"}

    except Exception as e:
        return {"type": "error", "result": str(e)}
````

````python
queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "Hey assistant, flip a coin for me!",
    "What time is it right now?",
    "What is machine learning?"
]
````

````python
for q in queries:
    print("Query:", q)
    print(json.dumps(agent(q), indent=4))
    print("-" * 50)
````
